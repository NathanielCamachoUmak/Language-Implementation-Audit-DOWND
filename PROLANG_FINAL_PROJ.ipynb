### 2.2 Performance & Memory Analysis
*Instructions: Pick two distinct implementations of the same logic and compare them using the code cells below.*

````python
# Metric 1: Execution Time (%timeit)
# Example: Comparing a list comprehension vs. a standard for-loop
import sys
import timeit

# Mock data simulating a list of member profiles from the database
members_data = [
    {"UserID": i, "StudName": f"Member {i}", "Rating": 1200 + (i % 500)}
    for i in range(1000)
]

def process_with_loop(data):
    """Traditional for-loop approach."""
    result = []
    for item in data:
        if item["Rating"] > 1400:
            result.append(item["StudName"])
    return result

def process_with_comprehension(data):
    """List comprehension approach (Pythonic equivalent of JS map/filter)."""
    return [item["StudName"] for item in data if item["Rating"] > 1400]

t1 = timeit.timeit(lambda: process_with_loop(members_data), number=1000)
t2 = timeit.timeit(lambda: process_with_comprehension(members_data), number=1000)

print(f"Execution Time (1000 runs):")
print(f"Standard Loop: {t1:.4f}s")
print(f"List Comprehension: {t2:.4f}s")
````

````python
# Metric 2: Memory Footprint (sys.getsizeof)
import sys

names_list = process_with_comprehension(members_data)
names_generator = (item["StudName"] for item in members_data if item["Rating"] > 1400)

print(f"\nMemory Footprint:")
print(f"Names List (Full list in memory): {sys.getsizeof(names_list)} bytes")
print(f"Names Generator (Lazy evaluation): {sys.getsizeof(names_generator)} bytes")
````

### Overall Conclusion

Moving from PHP to React isn't just a new tool, it's a completely different way of thinking about how apps work.
With a classic PHP/MySQL setup, everything happens in a neat, predictable sequence. User logs in, server checks the database, page reloads, done. It's simple to follow and easy to reason about. The downside? That constant full-page reload can feel pretty jarring compared to what people are used to today.
React and Supabase change the game entirely. Instead of the server doing all the heavy lifting, everything happens live in the browser, no reloads, no waiting, just that smooth, instant feel that modern apps are known for. Users love it.
The catch is that you, the developer, absorb all that complexity yourself. Authentication states, secure token handling, async data fetching, none of that is magic anymore, it's your problem to manage. The server used to quietly handle all of that. Now it's sitting right there in your codebase.
So yeah, the JavaScript stack is genuinely harder to architect. But when you get it right, you're building something that actually feels alive, and honestly, that's what people expect now.

1. **Paradigm & Language Shift:** If you had to rebuild this project in a significantly different modern programming language (e.g., Rust, Go, Swift, or Elixir), which of the 8 concepts would be the most difficult to translate or would require the most significant architectural change? Why?

If we decided to rebuild this whole system using a language like **Rust**, the biggest changes would be because of **Concept G: Concurrency** and **Concept A: Data Types (Memory Management)**.

* **Concurrency (Ownership & Borrowing):** Right now in our JavaScript code (like in the `dashboard.jsx` file), it is really easy to share data around while different things are loading at the same time. But Rust has very strict rules about which part of the code "owns" the data. We would have to be very careful and use specific, complicated tools like `Arc<Mutex<T>>` or `Channels` just to share things safely. The easygoing way we pass `adminData` around in React wouldn't work anymore; we'd need a much stricter way to share data so the program doesn't break.
* **Data Types & Memory:** JavaScript cleans up memory automatically and isn't super strict about data types, which lets us make flexible data objects like we did in `Members.jsx`. If we used C or Rust, we would have to create very strict `structs` (like rigid templates) for every single thing in the database. Also, because Rust doesn't clean up memory for you automatically like JavaScript does, we would have to manually keep track of exactly how long every single string of text from the database is allowed to stay in memory. This would make the part of the code that talks to the database a lot harder to write compared to the simple `supabase-js` we are using right now.</Mutex

2. **Language Evaluation:** Based on this audit, did the language(s) you chose help or hinder the development of your software engineering project? (Cite specific concepts like Concurrency or Exception Handling).

Choosing React and JavaScript was a big help in finishing this project quickly. This was mainly because of Concept G: Concurrency and Concept H: Event Handling.

* **Help (Concurrency):** By using the `async/await` method in the `login.js` file, doing the multi-step login and database requests became a lot simpler. We avoided having really confusing code, which let us write the "Bouncer" security logic using only a few lines.

If we used a language with strict typing like TypeScript or PHP, it would have pointed out these errors while we were coding instead of when we actually ran the program.

3. **Future Application:** How has auditing these concepts changed the way you will approach choosing a programming language for your next large-scale software project?

Looking at these 8 concepts changed our mind about how to pick a programming language. Instead of just picking "what is popular," We now try to look for **"what style actually fits the problem we are trying to solve."**

That way, the background logic doesn't get completely mixed up with the user interface code.

It showed me that if a project handles sensitive user data, we have to pick a language that has strong and easy-to-read ways to catch and handle errors.
